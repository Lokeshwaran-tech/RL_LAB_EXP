```python
import random

# Define the Environment
class Environment:
    def __init__(self):
        self.state = 0  # Agent starts at position 0

    def reset(self):
        self.state = 0  # Reset to starting position
        return self.state

    def step(self, action):
        # Update state based on action
        self.state += action

        # Rewards and conditions
        if self.state == 10:  # Treasure
            reward = 20
            done = True  # Episode ends
        elif self.state == 5:  # Trap
            reward = -5
            done = False
        else:
            reward = 0
            done = False

        return self.state, reward, done


# Define the Agent
class Agent:
    def __init__(self):
        self.actions = [-1, +1]  # Possible moves: backward (-1) or forward (+1)

    def select_action(self):
        return random.choice(self.actions)  # Random policy (no learning yet)


# Create Environment and Agent
env = Environment()
agent = Agent()

# Run one episode
state = env.reset()
done = False
step_count = 0

print("Starting Treasure Hunt...")

while not done:
    action = agent.select_action()
    next_state, reward, done = env.step(action)

    # Print step info
    print(f"Step {step_count}: State={state}, Action={action}, Next State={next_state}, Reward={reward}")

    state = next_state
    step_count += 1

print("Episode finished! Treasure found!" if state == 10 else "Episode finished!")
```
